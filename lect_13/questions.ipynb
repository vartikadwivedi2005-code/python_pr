```python
x = "Hello everyone lets learn python"
# output - Python learn lets everyone Hello
print(x)
words = x.split()
reversed_words = words[::-1]    #reversing the list of words
reversed_string = " ".join(reversed_words)      # joining the reversed list of words into a string
new_string = reversed_string.capitalize()
print(new_string)
```

```python
# Write a program to check if a given stringis a valid email address or not.
x= "exam.com"
# Check @ comes earlier than 'dot'
index=x.find("@")
dot_index=x.find(".")
if index != -1 and dot_index != -1 and index < dot_index:
    print("Valid email address")
else:
    print("Invalid email address")
print(index)
print(dot_index)
```
