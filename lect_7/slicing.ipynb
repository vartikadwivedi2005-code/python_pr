The standard syntax for a slice is:
sequence[start:stop:step]

start: The index where the slice begins (included). Defaults to 0.

stop: The index where the slice ends (excluded). Defaults to the end of the sequence.

step: The interval between elements. Defaults to 1.

```python
nums = [0, 1, 2, 3, 4, 5]
# nums[1:4]    [1, 2, 3]
# nums[:3]     [0, 1, 2]
# nums[3:]     [3, 4, 5]
# nums[0:6:2]  [0, 2, 4]
nums[:]
```
