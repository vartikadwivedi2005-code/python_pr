A tuple is one of Python's four built-in data types used to store collections of data. Think of it as a list that cannot be changed.

1. Key Characteristics
-> Ordered: They maintain the order in which elements are defined.

-> Immutable: Once a tuple is created, you cannot add, remove, or change its items.

-> Allow Duplicates: Since items are indexed, you can have multiple identical values.

-> Heterogeneous: You can mix different data types (strings, integers, booleans) in one tuple.

```python
# Empty tuple
tpl = ()
tpl1 = tuple()
print(tpl.tpl1)
```

```python
# Creating a tuple
fruits = ("apple", "banana", "cherry", "apple")

# Accessing by index
print(fruits[0])  # Output: apple

# Length of a tuple
print(len(fruits)) # Output: 4
```
