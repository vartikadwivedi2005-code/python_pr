stirng in python

A string in Python is a sequence of characters used to represent text. In Python, strings are immutable, meaning once you create one, you cannot change the characters inside it (though you can create a brand new string based on the old one).

```python
s1 = "learning"
s2 = "python"
print(s1)
# concatanation of 2 strings
s1 = s1 + s2
print(s1)
# repeat string n times
s = s1 * 4
print(s)
```

```python
print('o' in "hello world")
```

```python
x = "HELLO"
x = x.lower()
print(x)
```

```python
x = "PYTHON"
y = x.lower()
print(x)
print(y)
```

```python
# capitalize
a = "PYTHON I am Python".capitalize()
print(a)
```

```python
# split
str = "hello user i, am - learning, python"
lst=str.split()
print(lst)
```

```python
# split
str = "hello user i, am - learning, python"
lst=str.split(sep='a')
print(lst)
```

```python
words_list = ["hello","user","i","am", "learning","python"]
temp=" ".join(words_list)
print(temp)
```

```python
text = "apple banana apple cherry apple"
modified_text = text.replace("apple","orange")
print(text)
print(modified_text) 
```

```python
text = "apple,banana.apple,cherry,apple."
modified_text = text.replace(",",".").replace(".",",")
print(text)
print(modified_text)
```
