```python
# Wap grade students based on marks
marks = int(input("Enter marks:"))
if(marks>=90):
    grade="A"
elif(90>marks>=80):
    grade="B"
elif(80>marks>=70):
    grade="C"
elif(70>marks>=60):
    grade="D"
elif(60>marks>=50):
    grade="E"
else:
    grade="F"

print("Grade of students bases on marks:", grade)
```
