```python
# Conditional Statement in Python - if - elif - else - this is used to check multiple conditions and execute a block of code based on the first true condition. If none of the conditions are true then it will execute the block of code in else statement.
light = "red"
if(light == "red"):
    print("Stop")
elif(light == "yellow"):
    print("Get ready")
elif(light == "green"):
    print("Go")
else:
    print("Traffic light is not working")

print("This will always execute")
```
