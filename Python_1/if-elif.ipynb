```python
# Condtional Statements in Python - if - elif - this is used to check multiple conditions and execute a block of code based on the first true condition.
light = "green"
if light == "red":
    print("Stop")
elif light == "yellow":
    print("Get ready")
elif light == "green":
    print("Go")
print("This will always execute")
```

```python
# if-elif is only used when we have multiple conditions to check and we want to execute a block of code based on the first true condition. If we have only one condition to check then we can use if statement.
num =5
if(num>2):
    print("Number is greater than 2")
if(num>3):
    print("Number is greater than 3")
# difference between if and if-elif is that if-elif will check for the first true condition and execute the block of code based on that condition, while if will check for all the conditions and execute the block of code for all the true conditions.
```
