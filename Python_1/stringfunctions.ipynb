```python
# str.endswith() method in Python is used to check if a string ends with a specified suffix. It returns True if the string ends with the specified suffix, otherwise it returns False.
# returns true if string ends with substring, otherwise false
str = "I am a string"
print(str.endswith("string"))
print(str.endswith("am"))
```

```python
# str.capitalize() method in Python is used to convert the first character of a string to uppercase and the rest of the characters to lowercase. It returns a new string with the first character capitalized and the rest of the characters in lowercase.
# returns a string with the first character capitalized and the rest of the characters in lowercase
str ="i am a string"
str = str.capitalize()
# print(str.capitalize())
# print(str)
print(str)
```

```python
# str.replace() method in Python is used to replace occurrences of a specified substring with another substring. It returns a new string with the specified substring replaced by the new substring.
# returns a string where a specified substring is replaced with another substring
str = "I am a string"
str = str.replace("string", "number")
print(str)
```

```python
# str.find() method in Python is used to find the first occurrence of a specified substring in a string. It returns the index of the first occurrence of the substring, or -1 if the substring is not found.
# returns the index of the first occurrence of a specified substring, or -1 if the substring is not found
str = "I am a string"
print(str.find("string"))
print(str.find("number"))
# -1 is returned when the substring is not found in the string.
```

```python
# str.count() method in Python is used to count the number of occurrences of a specified substring in a string. It returns the number of occurrences of the substring in the string.
# returns the number of occurrences of a specified substring in a string
str = "I am a string. I am a string."
print(str.count("string"))
print(str.count("number"))
# 0 is returned when the substring is not found in the string.
```
