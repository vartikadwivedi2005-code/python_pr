```python
# Nesting conditions in Python - if statement inside another if statement is called nesting conditions. It is used to check multiple conditions and execute a block of code based on the true condition.

age =95
if(age >=18):  # if condition is true then it will execute the block of code
    if(age >=80):
        print("You are a senior citizen - cannot drive")
    else:
        print("You are an adult - can drive")
else:
    print("You are a minor - cannot drive")

print("Rules for driving...")
```
