```python
# WAP to input two numbers and print their sum, difference, product and quotient.

#Arithmetic operators - +, -, *, /, //, %, **
a = 5
b = 2
sum = a+b
print("Sum :", sum)
print("a+b :", a+b)   #addition
print("a-b :", a-b)   #substraction
print("a*b :", a*b)   #multiplication
print("a/b :", a/b)   #division   floating point number deta hai
print("a%b :", a%b)   # remainder
print("a**b :", a**b)  # exponentiation
print("a//b :", a//b)  # floor division   floating point number ko integer me convert kar deta hai  round off nahi karta h
```

```python
# assignment operators - =, +=, -=, *=, /=, //=, %=, **=
# #relational operators - ==, !=, >, <, >=, <=
a =50
b =20
print(a == b) #equal to
print(a != b) #not equal to
print(a>b)  #greater than
print(a<b)  #less than
print(a>=b) #greater than or equal to
print(a<=b) #less than or equal to
```

```python
# logical operators - logical operators are used to combine multiple conditions and return a boolean value based on the result of those conditions. The three main logical operators in Python are "and", "or", and "not".
#logical operators  - and, or, not
print(not True)
print(not False)
a = 40
b = 20
print(not (a>b)) #False

val1 = True
val2 = True
print("AND operator:",val1 and val2) 

val1 = True
val2 = False
print("AND operator:",val1 and val2) 

val1 = True
val2 = False
print("OR operator:",val1 or val2)


val1 = False
val2 = False
print("OR operator:",val1 or val2) 

print("OR operator:" , (a == b) or (a > b)) #True


```
