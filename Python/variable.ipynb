```python
# Variable - a variable is a named location in memory that stores a value. It can be used to store data of different types, such as numbers, strings, lists, etc. In Python, you can create a variable by simply assigning a value to it using the assignment operator (=). The variable name should follow certain rules, such as it should start with a letter or an underscore, and it can contain letters, numbers, and underscores. Variables are case-sensitive in Python, which means that 'name' and 'Name' are considered different variables.

name = "Vartika"
age = 20
price = 20.44
print(name)
print(age)
print(price)
age2 = age
print("My name is : ",name)
print("My age is :",age)
print(age2)
print(type(name))
print(type(age))
print(type(price))
```
