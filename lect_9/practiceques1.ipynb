Wap to program using a list comprehension to create a list of even numbers less than 8 from 1 to 20.

```python

numbers = range(1, 21)

# List comprehension with two conditions:
# 1. i % 2 == 0 (must be even)
# 2. i < 8 (must be less than 8)
even_num = [i for i in numbers if i % 2 == 0 and i < 8]

print(even_num)
```

Wap using a list comprehension to create a list of numbers from 1 to 10 , but only include numbers that are greater than 5.

Wap using a list comprehension to create a list of the first letters of each word in the given list.
list = ['apple','banana','cherry']



```python
# range(1, 11) generates numbers from 1 to 10
list = [i for i in range(1, 11) if i > 5]

print(list)
```

```python
words = ['apple', 'banana', 'cherry', 'apricot']

# Syntax: [expression_if_true if condition else expression_if_false for item in list]
processed_words = [word if word.startswith('a') else word.upper() for word in words]

print(f"Original:  {words}")
print(f"Processed: {processed_words}")
```
