```python
# Conditiona; statements with practice questions
#Conditional Statements - if, else, elif
age = int (input("Enter your age:"))
if age >=18:
    print("ELIGIBLE TO VOTE")
else:
    print("NOT ELIGIBLE TO VOTE")

#INDENTATION - if, else, elif ke andar jo code hota hai usko indent karna padta hai. Indentation se python ko pata chalta hai ki kaunsa code kis statement ke andar hai. Indentation ke liye hum space ya tab ka use kar sakte hain. Python me indentation bahut important hai, agar indentation sahi nahi hoga to code error dega.

```

```python
num = int(input("Enter a number:"))
if num > 0:
    print ("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
num = int(input("Enter a number"))
if num % 3 == 0 and num % 5== 0:
    print("The number is divisible by both 3 and 5.")
elif num % 3 ==0:
    print("The number is divisible by 3.")
elif num % 5 == 0:
    print("The number is divisible by 5.")
else:
    print("The number is not divisible by any of the two")
```

```python
year = int(input("Enter a year:"))
if ( year % 4 == 0 and year % 100 !=0) or (year % 400 == 0):
    print("The year is a leap year.")
else:
    print("The year is not a leap year.")
```

```python
month = int(input("Enter a month(1-12):"))
if month >= 3 and month <= 5:
    print("The month is in spring.")
elif month >= 6 and month <= 8:
    print("The month is in summer.")
elif month >= 9 and month <= 11:
    print("The month is in autumn.")
else:
    print("The month is in winter.")
```
