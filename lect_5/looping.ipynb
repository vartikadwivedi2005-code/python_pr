Wap which will print FIZZ if the no is divisible by 3 BUZZ if the no is divisible by 5 and FIZZBUZZ if the no is divisible by both 3 and 5 otherwise 

```python
num = int(input("Enter a number:"))
# for i in range(1,30,3):
if(num % 3 == 0):
    print("FIZZ")
elif (num % 5 == 0):
    print("BUZZ")
else:
    print("FIZZBUZZ")
```

```python
num = int(input("Enter number:"))
for i in range(1,num+1):
    if(i % 3 == 0 and i % 5 == 0):
        print("FIZZBUZZ")
    elif(i % 3 == 0):
        print("FIZZ")
    elif(i % 5 == 0):
        print("BUZZ")
    else:
        print(i)
```

```python
# Wap by taking a number from user and print its factors.
num = int(input("Enter a number:"))
even = 0
odd = 0
for i in range(1,num+1):
    if(num % i == 0):
        if(i % 2 == 0):
             even += 1
        else:
             odd += 1
    print("Even factors:",even,"Odd factors:",odd)
```

```python
num = int(input("Enter a number:"))
for i in range(2,num//2):
  if(num % i == 0):
        print("No is not prime.")
        break
else:
        print("No is prime")
```

while vs for
Use a for loop when you know exactly how many times you want to repeat (like counting to 10).

Use a while loop when you want to keep going until something specific happens (like waiting for a user to type "quit").
