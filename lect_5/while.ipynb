A while loop runs as long as a specific condition stays True.
To make a while loop work without crashing your program, remember these three steps:

Initialization: Set your variable (like i = 1) before the loop starts.

The Condition: The loop checks while (condition): before every run.

Update: You must change the variable inside the loop (e.g., i = i + 1) so that the condition eventually becomes False.
