```python
list = [1,2,3,5]
list.clear()
print(list)

# clear or deletes the whole list.
```

```python
list = [1,2,6,8,6,4,9]
list.reverse()
print(list)

# reverse the given list.
```

```python
list = [3,6,8,5,9,5,8,4]
list.sort()
print(list)

# in place sort method 
```

```python
list = [3,6,8,5,9,5,8,4]
list.sort(reverse=True)
print(list)

# to print list in descending order
```

```python
# another method of sorting
# sorted is an in-built function of python.
list = [3,6,8,5,9,5,8,4]
list1 = sorted(list)
print(list)
print(list1)
```

```python
list = [3,6,8,5,9,5,8,4]
list1 = sorted(list,reverse=True)
print(list)
print(list1)
```
