```python
list = [4, 6, 5, 7, 4, 8]

list1 = []
for i in list:
    if i%2==0:
        list1.append(i)
print(list1)
```

```python
# List comprehension
list = [4, 6, 5, 7, 4, 8]
list1 = [i for i in list if i%2==0]
print(list1)
```

```python
list = [4, 6, 5, 7, 4, 8]
list1 = [i for i in list if i%2!=0]
print(list1)
```
